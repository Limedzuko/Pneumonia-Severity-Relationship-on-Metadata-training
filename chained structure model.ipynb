```python
# %%
# =============================================================================
# 1. IMPORTS AND SETUP
# =============================================================================
import os
import sys
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import seaborn as sns
from PIL import Image
import matplotlib.pyplot as plt
import random
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_score, recall_score, accuracy_score, f1_score
from sklearn.model_selection import train_test_split
from tqdm.notebook import tqdm
import torch.nn.functional as F

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)

# Setup device (GPU or CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")


def visualize_predictions(dataset_loader, model=None, num_samples=1, threshold=0.5):
    """
    Function to visualize images, ground truth masks, and predicted masks using a PyTorch DataLoader.
    """
    model.eval()  # Set the model to evaluation mode

    # Get one batch from the data loader
    (images, variables), (masks, targets, severities) = next(iter(dataset_loader))

    # Move data to CPU for visualization
    images, variables, masks, targets, severities = images.cpu(), variables.cpu(), masks.cpu(), targets.cpu(), severities.cpu()

    with torch.no_grad():  # Disable gradient calculation for inference
        for i in range(min(num_samples, len(images))):
            img_to_show = images[i]
            var_to_pred = variables[i:i+1].to(device) # Keep on device for model
            img_to_pred = img_to_show.unsqueeze(0).to(device) # Add batch dim and send to device

            show_pred = False
            if model is not None:
                pred_mask, pred_target, pred_severity = model(img_to_pred, var_to_pred)
                pred_mask = pred_mask.cpu() # Move prediction to CPU
                pred_target = pred_target.cpu()
                pred_severity = pred_severity.cpu()
                binarized_mask = (torch.sigmoid(pred_mask[0]) > threshold).float()
                show_pred = True

            # Convert image tensor for plotting (C, H, W) -> (H, W, C)
            img_display = img_to_show.permute(1, 2, 0).numpy()

            plt.figure(figsize=(10, 6))

            # Original image
            plt.subplot(1, 3 if show_pred else 2, 1)
            plt.imshow(img_display)
            plt.title('Original Image')
            plt.axis('off')

            # Ground truth mask
            plt.subplot(1, 3 if show_pred else 2, 2)
            plt.imshow(masks[i].squeeze(), cmap='gray')
            plt.title('Ground Truth Mask')
            plt.axis('off')

            # Predicted binarized mask
            if show_pred:
                plt.subplot(1, 3, 3)
                plt.imshow(binarized_mask.squeeze(), cmap='gray')
                plt.title('Predicted Mask (Binarized)')
                plt.axis('off')

            plt.tight_layout()
            plt.show()

            # Show the predicted binary target value and severity
            if model is not None:
                gt_target = targets[i].item()
                pred_target_class = (torch.sigmoid(pred_target[0]).item() > 0.5)
                pred_severity_score = torch.argmax(pred_severity[0]).item()
                print(f"Ground Truth Target: {gt_target}, Predicted Target: {int(pred_target_class)}, Predicted Severity: {pred_severity_score}")
            else:
                print(f"Ground Truth Target: {targets[i].item()}, No model predictions available.")


def plot_learning_curves(history):
    """
    Plots learning curves from a history dictionary.
    """
    plt.figure(figsize=(18, 5))

    # Loss
    plt.subplot(1, 3, 1)
    plt.plot(history['train_total_loss'], label='Training Total Loss')
    plt.plot(history['val_total_loss'], label='Validation Total Loss')
    plt.title('Total Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Accuracy
    plt.subplot(1, 3, 2)
    plt.plot(history['train_acc'], label='Training Target Accuracy')
    plt.plot(history['val_acc'], label='Validation Target Accuracy')
    plt.plot(history['train_severity_acc'], label='Training Severity Accuracy')
    plt.plot(history['val_severity_acc'], label='Validation Severity Accuracy')
    plt.title('Accuracy (Target & Severity)')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.ylim(0, 1)

    # IoU
    plt.subplot(1, 3, 3)
    plt.plot(history['train_iou'], label='Training IoU')
    plt.plot(history['val_iou'], label='Validation IoU')
    plt.title('Mask IoU')
    plt.xlabel('Epochs')
    plt.ylabel('IoU')
    plt.legend()
    plt.ylim(0, 1)


    plt.tight_layout()
    plt.show()


df = pd.read_csv('/kaggle/input/rsna-pneumonia-processed-dataset/stage2_train_metadata.csv')
df=df[df['class'].isin(["Lung Opacity","Normal"])].copy()
df.drop(['x', 'y', 'width', 'height','modality','class'],axis=1,inplace=True )
df= df.drop_duplicates()

df_train_val, df_test = train_test_split(df,test_size=0.1,stratify=df["Target"], random_state=42)
df_train, df_val = train_test_split(df_train_val,test_size=0.2,stratify=df_train_val["Target"], random_state=42)

sex_mapping = {'M': 0, 'F': 1}
position_mapping = {'AP': 0, 'PA': 1}

for d in [df_train, df_val, df_test]:
    d['sex'] = d['sex'].map(sex_mapping)
    d['position'] = d['position'].map(position_mapping)
    d['age'] = d['age'].apply(lambda x: 90 if x > 90 else x)

scaler = MinMaxScaler()
df_train['age'] = scaler.fit_transform(df_train[['age']])
df_val['age'] = scaler.transform(df_val[['age']])
df_test['age'] = scaler.transform(df_test[['age']])

print(f"Train set shape: {df_train.shape}")
print(f"Validation set shape: {df_val.shape}")
print(f"Test set shape: {df_test.shape}")


class PneumoniaDataset(Dataset):
    def __init__(self, df, images_dir, masks_dir, augmentation=False):
        self.df = df
        self.images_dir = images_dir
        self.masks_dir = masks_dir
        self.augmentation = augmentation

        self.image_paths = images_dir + df["patientId"].astype(str) + ".png"
        self.mask_paths = masks_dir + df["patientId"].astype(str) + ".png"
        self.variables_num = df[["age", "sex", "position"]].values
        self.targets = df['Target'].values

        self.resize_transform = transforms.Resize((224, 224), antialias=True)
        self.base_transforms = transforms.Compose([
            transforms.ToTensor(), # This also scales images to [0, 1]
        ])

        if self.augmentation:
            self.aug_transforms = transforms.Compose([
                transforms.ColorJitter(brightness=0.1, contrast=0.1),
            ])

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # Load image and mask
        image = Image.open(self.image_paths.iloc[idx]).convert("RGB")
        mask = Image.open(self.mask_paths.iloc[idx]).convert("L") # Grayscale

        # Resize
        image = self.resize_transform(image)
        mask = self.resize_transform(mask)

        # Apply augmentations if enabled
        if self.augmentation:
            image = self.aug_transforms(image)

        # Convert to tensor and normalize
        image = self.base_transforms(image)
        mask = self.base_transforms(mask)
        mask = (mask > 0).float() # Binarize mask

        # Get numerical variables and target
        variables = torch.tensor(self.variables_num[idx], dtype=torch.float32)
        target = torch.tensor(self.targets[idx], dtype=torch.float32)

        # For severity, we need the original mask size.
        # This is a placeholder, in a real scenario, you would calculate this
        # from the original bounding box data or the mask itself before resizing.
        # Here we'll use the target value as a proxy for simplicity in this example.
        # A more accurate approach would involve loading the original mask image
        # and calculating the percentage of white pixels within the estimated lung area.
        original_mask = Image.open(self.mask_paths.iloc[idx]).convert("L")
        # Estimate lung area as 35% of image area
        estimated_lung_area = original_mask.width * original_mask.height * 0.35
        mask_area = np.sum(np.array(original_mask) > 0)
        mask_percentage = (mask_area / estimated_lung_area) * 100 if estimated_lung_area > 0 else 0.0

        # Map mask percentage to severity score (0-5)
        if mask_percentage == 0:
            severity_score = 0
        elif mask_percentage < 5:
            severity_score = 1
        elif mask_percentage < 25:
            severity_score = 2
        elif mask_percentage < 49:
            severity_score = 3
        elif mask_percentage < 75:
            severity_score = 4
        else:
            severity_score = 5

        severity = torch.tensor(severity_score, dtype=torch.long)


        return (image, variables), (mask, target, severity)

# Directories
input_dir = "/kaggle/input/rsna-pneumonia-processed-dataset/"
images_dir = os.path.join(input_dir, "Training", 'Images/')
mask_dir = os.path.join(input_dir, "Training", 'Masks/')

# Create Datasets
train_dataset = PneumoniaDataset(df_train, images_dir, mask_dir, augmentation=True)
val_dataset = PneumoniaDataset(df_val, images_dir, mask_dir, augmentation=False)
test_dataset = PneumoniaDataset(df_test, images_dir, mask_dir, augmentation=False)

# Create DataLoaders
batch_size = 8
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

# Verify a batch
(images, variables), (masks, targets, severities) = next(iter(train_loader))
print("Images batch shape:", images.shape)
print("Variables batch shape:", variables.shape)
print("Masks batch shape:", masks.shape)
print("Targets batch shape:", targets.shape)
print("Severities batch shape:", severities.shape)



class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(ResidualBlock, self).__init__()
        self.bn1 = nn.BatchNorm2d(in_channels)
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding='same', bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding='same', bias=False)
        self.bn3 = nn.BatchNorm2d(out_channels)
        self.leaky_relu = nn.LeakyReLU(negative_slope=0.1, inplace=True)

        # Skip connection handling
        self.skip_conv = None
        if in_channels != out_channels:
            self.skip_conv = nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=1, bias=False)

    def forward(self, x):
        identity = x

        out = self.bn1(x)
        out = self.conv1(out)
        out = self.bn2(out)
        out = self.leaky_relu(out)
        out = self.conv2(out)
        out = self.bn3(out)

        if self.skip_conv:
            identity = self.skip_conv(identity)

        out += identity
        out = self.leaky_relu(out)
        return out

class UnetHybrid(nn.Module):
    def __init__(self, image_size=224):
        super(UnetHybrid, self).__init__()
        self.image_size = image_size

        # Encoder
        self.conv_in = nn.Sequential(nn.Conv2d(3, 32, 3, padding='same', bias=False), nn.BatchNorm2d(32), nn.LeakyReLU(0.1, inplace=True))
        self.enc1 = nn.Sequential(nn.Conv2d(32, 32, 3, padding='same', bias=False), nn.BatchNorm2d(32), nn.LeakyReLU(0.1, inplace=True))
        self.pool1 = nn.MaxPool2d(2)
        self.res1 = ResidualBlock(32, 32)

        self.enc2 = nn.Sequential(nn.Conv2d(32, 64, 3, padding='same', bias=False), nn.BatchNorm2d(64), nn.LeakyReLU(0.1, inplace=True))
        self.pool2 = nn.MaxPool2d(2)
        self.res2 = ResidualBlock(64, 64)

        self.enc3 = nn.Sequential(nn.Conv2d(64, 128, 3, padding='same', bias=False), nn.BatchNorm2d(128), nn.LeakyReLU(0.1, inplace=True))
        self.pool3 = nn.MaxPool2d(2)
        self.res3 = ResidualBlock(128, 128)

        self.enc4 = nn.Sequential(nn.Conv2d(128, 256, 3, padding='same', bias=False), nn.BatchNorm2d(256), nn.LeakyReLU(0.1, inplace=True))
        self.pool4 = nn.MaxPool2d(2)
        self.res4 = ResidualBlock(256, 256)

        # Bottleneck
        self.bottleneck = nn.Sequential(
            nn.Conv2d(256, 512, 3, padding='same', bias=False), nn.BatchNorm2d(512), nn.LeakyReLU(0.1, inplace=True),
            nn.Conv2d(512, 512, 3, padding='same', bias=False), nn.BatchNorm2d(512), nn.LeakyReLU(0.1, inplace=True),
            nn.Dropout(0.3)
        )

        # Decoder
        self.up6 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.dec_res6 = ResidualBlock(512, 256)
        self.up7 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.dec_res7 = ResidualBlock(256, 128)
        self.up8 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.dec_res8 = ResidualBlock(128, 64)
        self.up9 = nn.ConvTranspose2d(64, 32, 2, stride=2)
        self.dec_res9 = ResidualBlock(64, 32)

        # Numerical variable processing
        self.var_dense = nn.Sequential(nn.Linear(3, 64), nn.ReLU(), nn.BatchNorm1d(64))

        # Output layers
        self.mask_output = nn.Conv2d(32 + 64, 1, kernel_size=1) # 32 from decoder, 64 from vars

        self.classifier_head = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(32 + 64, 64, bias=False),
            nn.BatchNorm1d(64),
            nn.LeakyReLU(0.1, inplace=True),
            nn.Linear(64, 1)
        )

        # Layers to process mask and classifier outputs for severity head
        # Process mask output (from spatial to a fixed-size vector)
        self.mask_severity_processor = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)), # Pool spatial mask output to 1x1
            nn.Flatten(), # Flatten to a vector (size 1)
            nn.Linear(1, 16), # Project to a small dimension
            nn.LeakyReLU(0.1, inplace=True),
            nn.BatchNorm1d(16)
        )

        # Process classifier output (from 1 to a small vector)
        self.classifier_severity_processor = nn.Sequential(
            nn.Linear(1, 16), # Project to a small dimension
            nn.LeakyReLU(0.1, inplace=True),
            nn.BatchNorm1d(16)
        )

        # Process merged_output for severity head
        self.merged_output_severity_processor = nn.Sequential(
             nn.AdaptiveAvgPool2d((1, 1)), # Pool merged_output features
             nn.Flatten() # Flatten to a vector (size 32 + 64)
        )


        # The input size to severity_head will be (32+64) from pooled merged_output
        # + 16 from processed mask output + 16 from processed classifier output
        severity_input_size = (32 + 64) + 16 + 16 # 96 + 16 + 16 = 128

        self.severity_head = nn.Sequential(
            nn.Linear(severity_input_size, 128, bias=False),
            nn.BatchNorm1d(128),
            nn.LeakyReLU(0.1, inplace=True),
            nn.Dropout(0.3),
            nn.Linear(128, 6) # 6 output classes for severity 0-5
        )


    def forward(self, img, var):
        # Encoder
        c0 = self.conv_in(img)
        c1 = self.enc1(c0)
        p1 = self.pool1(c1)
        r1 = self.res1(p1)

        c2 = self.enc2(r1)
        p2 = self.pool2(c2)
        r2 = self.res2(p2)

        c3 = self.enc3(r2)
        p3 = self.pool3(c3)
        r3 = self.res3(p3)

        c4 = self.enc4(r3)
        p4 = self.pool4(c4)
        r4 = self.res4(p4)

        # Bottleneck
        neck = self.bottleneck(r4)

        # Decoder
        u6 = self.up6(neck)
        # Ensure spatial dimensions match before concatenation
        if u6.shape[-2:] != c4.shape[-2:]:
             u6 = F.interpolate(u6, size=c4.shape[-2:], mode='nearest')
        m6 = torch.cat([c4, u6], dim=1)
        dr6 = self.dec_res6(m6)

        u7 = self.up7(dr6)
        if u7.shape[-2:] != c3.shape[-2:]:
             u7 = F.interpolate(u7, size=c3.shape[-2:], mode='nearest')
        m7 = torch.cat([c3, u7], dim=1)
        dr7 = self.dec_res7(m7)

        u8 = self.up8(dr7)
        if u8.shape[-2:] != c2.shape[-2:]:
             u8 = F.interpolate(u8, size=c2.shape[-2:], mode='nearest')
        m8 = torch.cat([c2, u8], dim=1)
        dr8 = self.dec_res8(m8)

        u9 = self.up9(dr8)
        if u9.shape[-2:] != c1.shape[-2:]:
             u9 = F.interpolate(u9, size=c1.shape[-2:], mode='nearest')
        m9 = torch.cat([c1, u9], dim=1)
        dr9 = self.dec_res9(m9)

        # Process and merge numeric variables
        var_proc = self.var_dense(var)
        var_reshaped = var_proc.view(-1, 64, 1, 1)
        # Upsample numerical variables to match decoder output size
        var_upsampled = F.interpolate(var_reshaped, size=dr9.shape[-2:], mode='nearest')


        merged_output = torch.cat([dr9, var_upsampled], dim=1)

        # Primary outputs from merged_output
        mask_out = self.mask_output(merged_output)
        target_out = self.classifier_head(merged_output)

        # Process mask and classifier outputs for severity head
        processed_mask_out = self.mask_severity_processor(mask_out)
        # Ensure target_out has shape (batch_size, 1) before processing
        processed_target_out = self.classifier_severity_processor(target_out.view(target_out.size(0), 1))


        # Process merged_output for severity head
        processed_merged_output = self.merged_output_severity_processor(merged_output)

        # Concatenate all processed inputs for severity head
        severity_input = torch.cat([processed_merged_output, processed_mask_out, processed_target_out], dim=1)

        # Final severity output
        severity_out = self.severity_head(severity_input)


        return mask_out, target_out, severity_out

model = UnetHybrid().to(device)

mask_loss_fn = nn.BCEWithLogitsLoss()

y = df_train["Target"].values
class_weights = len(y) / (2 * np.bincount(y))
class_weights_tensor = torch.tensor([class_weights[1]], dtype=torch.float32).to(device) # pos_weight for class 1
target_loss_fn = nn.BCEWithLogitsLoss(pos_weight=class_weights_tensor)

# Loss function for ordinal severity (Cross-Entropy for classification)
severity_loss_fn = nn.CrossEntropyLoss()


def iou_loss(pred, target):
    pred = torch.sigmoid(pred)
    intersection = (pred * target).sum()
    union = pred.sum() + target.sum() - intersection
    iou = (intersection + 1e-6) / (union + 1e-6)
    return 1 - iou

def iou_metric(pred, target, threshold=0.5):
    pred = torch.sigmoid(pred) > threshold
    target = target > 0.5
    intersection = (pred & target).sum().float()
    union = (pred | target).sum().float()
    return (intersection + 1e-6) / (union + 1e-6)


learning_rate = 0.0001
epochs = 20
patience = 5
output_dir = "/kaggle/working/"

# Optimizer and Scheduler
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.2, patience=2)

history = {
    'train_total_loss': [], 'val_total_loss': [],
    'train_mask_loss': [], 'val_mask_loss': [],
    'train_target_loss': [], 'val_target_loss': [],
    'train_severity_loss': [], 'val_severity_loss': [],
    'train_iou': [], 'val_iou': [],
    'train_acc': [], 'val_acc': [],
    'train_severity_acc': [], 'val_severity_acc': []
}
best_val_loss = float('inf')
patience_counter = 0

for epoch in range(epochs):
    model.train()
    train_total_loss, train_mask_loss, train_target_loss, train_severity_loss = 0, 0, 0, 0
    train_iou, train_correct_target, train_total_target, train_correct_severity, train_total_severity = 0, 0, 0, 0, 0


    train_pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} [Train]", leave=False)
    for (images, variables), (masks, targets, severities) in train_pbar:
        images, variables, masks, targets, severities = images.to(device), variables.to(device), masks.to(device), targets.to(device), severities.to(device)

        pred_mask, pred_target, pred_severity = model(images, variables)

        loss_m = iou_loss(pred_mask, masks)
        loss_t = target_loss_fn(pred_target.squeeze(), targets)
        loss_s = severity_loss_fn(pred_severity, severities)

        # Combine losses with weights (adjust weights as needed)
        total_loss = 0.5 * loss_m + 0.3 * loss_t + 0.2 * loss_s

        optimizer.zero_grad()
        total_loss.backward()
        optimizer.step()

        train_total_loss += total_loss.item()
        train_mask_loss += loss_m.item()
        train_target_loss += loss_t.item()
        train_severity_loss += loss_s.item()

        train_iou += iou_metric(pred_mask, masks).item()

        predicted_target_classes = (torch.sigmoid(pred_target).squeeze() > 0.5)
        train_correct_target += (predicted_target_classes == targets).sum().item()
        train_total_target += targets.size(0)

        predicted_severity_classes = torch.argmax(pred_severity, dim=1)
        train_correct_severity += (predicted_severity_classes == severities).sum().item()
        train_total_severity += severities.size(0)

        train_pbar.set_postfix({'loss': total_loss.item()})

    model.eval()
    val_total_loss, val_mask_loss, val_target_loss, val_severity_loss = 0, 0, 0, 0
    val_iou, val_correct_target, val_total_target, val_correct_severity, val_total_severity = 0, 0, 0, 0, 0


    val_pbar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{epochs} [Val]", leave=False)
    with torch.no_grad():
        for (images, variables), (masks, targets, severities) in val_pbar:
            images, variables, masks, targets, severities = images.to(device), variables.to(device), masks.to(device), targets.to(device), severities.to(device)
            pred_mask, pred_target, pred_severity = model(images, variables)

            loss_m = iou_loss(pred_mask, masks)
            loss_t = target_loss_fn(pred_target.squeeze(), targets)
            loss_s = severity_loss_fn(pred_severity, severities)

            total_loss = 0.5 * loss_m + 0.3 * loss_t + 0.2 * loss_s

            val_total_loss += total_loss.item()
            val_mask_loss += loss_m.item()
            val_target_loss += loss_t.item()
            val_severity_loss += loss_s.item()

            val_iou += iou_metric(pred_mask, masks).item()

            predicted_target_classes = (torch.sigmoid(pred_target).squeeze() > 0.5)
            val_correct_target += (predicted_target_classes == targets).sum().item()
            val_total_target += targets.size(0)

            predicted_severity_classes = torch.argmax(pred_severity, dim=1)
            val_correct_severity += (predicted_severity_classes == severities).sum().item()
            val_total_severity += severities.size(0)


    avg_train_total_loss = train_total_loss / len(train_loader)
    avg_train_mask_loss = train_mask_loss / len(train_loader)
    avg_train_target_loss = train_target_loss / len(train_loader)
    avg_train_severity_loss = train_severity_loss / len(train_loader)
    avg_train_iou = train_iou / len(train_loader)
    avg_train_acc_target = train_correct_target / train_total_target
    avg_train_acc_severity = train_correct_severity / train_total_severity


    avg_val_total_loss = val_total_loss / len(val_loader)
    avg_val_mask_loss = val_mask_loss / len(val_loader)
    avg_val_target_loss = val_target_loss / len(val_loader)
    avg_val_severity_loss = val_severity_loss / len(val_loader)
    avg_val_iou = val_iou / len(val_loader)
    avg_val_acc_target = val_correct_target / val_total_target
    avg_val_acc_severity = val_correct_severity / val_total_severity


    history['train_total_loss'].append(avg_train_total_loss)
    history['train_mask_loss'].append(avg_train_mask_loss)
    history['train_target_loss'].append(avg_train_target_loss)
    history['train_severity_loss'].append(avg_train_severity_loss)
    history['train_iou'].append(avg_train_iou)
    history['train_acc'].append(avg_train_acc_target)
    history['train_severity_acc'].append(avg_train_acc_severity)

    history['val_total_loss'].append(avg_val_total_loss)
    history['val_mask_loss'].append(avg_val_mask_loss)
    history['val_target_loss'].append(avg_val_target_loss)
    history['val_severity_loss'].append(avg_val_severity_loss)
    history['val_iou'].append(avg_val_iou)
    history['val_acc'].append(avg_val_acc_target)
    history['val_severity_acc'].append(avg_val_acc_severity)


    print(f"Epoch {epoch+1}/{epochs} -> Train Loss (Total): {avg_train_total_loss:.4f}, Train Acc (Target): {avg_train_acc_target:.4f}, Train IoU: {avg_train_iou:.4f}, Train Acc (Severity): {avg_train_acc_severity:.4f} | "
          f"Val Loss (Total): {avg_val_total_loss:.4f}, Val Acc (Target): {avg_val_acc_target:.4f}, Val IoU: {avg_val_iou:.4f}, Val Acc (Severity): {avg_val_acc_severity:.4f}")

    scheduler.step(avg_val_total_loss)

    if avg_val_total_loss < best_val_loss:
        best_val_loss = avg_val_total_loss
        torch.save(model.state_dict(), os.path.join(output_dir, 'model_unet_hybrid_severity.pth'))
        print("Model saved as validation loss improved.")
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1}.")
            break

history_df = pd.DataFrame(history)
history_df.to_csv(os.path.join(output_dir, 'history_unet_hybrid_severity.csv'), index=False)

# Load the best model weights for evaluation
model.load_state_dict(torch.load(os.path.join(output_dir, 'model_unet_hybrid_severity.pth')))
model.eval()

# Visualize learning curves
plot_learning_curves(history)

# Evaluate on test set and collect predictions for metrics
all_targets = []
all_preds = []
all_severities_gt = []
all_severities_pred = []
all_mask_ious = []

with torch.no_grad():
    for (images, variables), (masks, targets, severities) in tqdm(test_loader, desc="Evaluating on test set"):
        images, variables, masks, targets, severities = images.to(device), variables.to(device), masks.to(device), targets.to(device), severities.to(device)

        pred_mask, pred_target, pred_severity = model(images, variables)

        # Collect data for classification metrics
        predicted_target_classes = (torch.sigmoid(pred_target).squeeze() > 0.5).cpu().numpy()
        all_targets.extend(targets.cpu().numpy())
        all_preds.extend(predicted_target_classes)

        # Collect data for severity metrics
        predicted_severity_classes = torch.argmax(pred_severity, dim=1).cpu().numpy()
        all_severities_gt.extend(severities.cpu().numpy())
        all_severities_pred.extend(predicted_severity_classes)

        # Collect data for mask IoU
        batch_ious = iou_metric(pred_mask, masks, threshold=0.5)
        all_mask_ious.append(batch_ious.item())

# Calculate and print classification metrics
print("\n" + "="*20 + " TEST SET CLASSIFICATION METRICS (Target) " + "="*20)
print(classification_report(all_targets, all_preds))
print(f"Overall Accuracy: {accuracy_score(all_targets, all_preds):.4f}")
print(f"Overall F1 Score: {f1_score(all_targets, all_preds):.4f}")

# Plot Confusion Matrix for Target
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(all_targets, all_preds), annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Pneumonia'], yticklabels=['Normal', 'Pneumonia'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Target)')
plt.show()


# Calculate and print severity accuracy and confusion matrix
print("\n" + "="*20 + " TEST SET SEVERITY METRICS " + "="*20)
print(f"Severity Classification Report:\n{classification_report(all_severities_gt, all_severities_pred)}")
print(f"Severity Accuracy: {accuracy_score(all_severities_gt, all_severities_pred):.4f}")

# Plot Confusion Matrix for Severity
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(all_severities_gt, all_severities_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=range(6), yticklabels=range(6))
plt.xlabel('Predicted Severity')
plt.ylabel('Actual Severity')
plt.title('Confusion Matrix (Severity)')
plt.show()


# Calculate and print average Mask IoU
average_iou = np.mean(all_mask_ious)
print(f"\nAverage Mask IoU on Test Set: {average_iou:.4f}")

# Visualize predictions with severity
severity_levels_to_show = list(range(6)) # Severity levels 0 through 5
found_severities = {level: False for level in severity_levels_to_show}

print("\nVisualizing one example image, ground truth mask, and predicted mask per severity class:")

model.eval() # Ensure model is in evaluation mode

# Iterate through the test loader to find one example for each severity level
with torch.no_grad():
    for (images, variables), (masks, targets, severities) in test_loader:
        # Move data to device for model prediction, and to CPU for visualization
        images_device = images.to(device)
        variables_device = variables.to(device)
        masks_cpu = masks.cpu()
        targets_cpu = targets.cpu()
        images_cpu = images.cpu()
        severities_cpu = severities.cpu()

        # Get model predictions for the batch
        pred_mask, pred_target, pred_severity = model(images_device, variables_device)

        # Move predictions to CPU for processing and visualization
        pred_mask_cpu = pred_mask.cpu()
        pred_target_cpu = pred_target.cpu()
        pred_severity_cpu = pred_severity.cpu()


        for i in range(len(images_cpu)):
            current_severity = severities_cpu[i].item()
            current_target = targets_cpu[i].item()

            if current_severity in severity_levels_to_show and not found_severities[current_severity]:
                img_to_show = images_cpu[i]
                mask_to_show = masks_cpu[i]
                pred_mask_to_show = (torch.sigmoid(pred_mask_cpu[i]) > 0.5).float()

                # Recalculate mask percentage for display purposes (optional, but good for verification)
                # This assumes you have access to the original mask path or can reconstruct it
                # For this example, we'll use the severity value which is derived from mask percentage
                # In a real scenario, you'd load the original mask here
                # Placeholder: Get an estimated percentage based on the severity score
                severity_percentage_map = {
                    0: "0%",
                    1: "<5%",
                    2: "5-25%",
                    3: "26-49%",
                    4: "50-75%",
                    5: ">75%"
                }
                percentage_text = severity_percentage_map.get(current_severity, "N/A")

                # Map target to label
                target_label = "Pneumonia" if current_target == 1 else "Normal"

                # Get predicted binary target and severity
                predicted_target_class = (torch.sigmoid(pred_target_cpu[i]).item() > 0.5)
                predicted_severity_score = torch.argmax(pred_severity_cpu[i]).item()
                predicted_target_label = "Pneumonia" if predicted_target_class else "Normal"


                plt.figure(figsize=(12, 5))

                # Original image
                plt.subplot(1, 3, 1)
                plt.imshow(img_to_show.permute(1, 2, 0).numpy()) # Convert tensor for plotting (C, H, W) -> (H, W, C)
                plt.title(f'Severity {current_severity} - Image')
                plt.axis('off')

                # Ground truth mask
                plt.subplot(1, 3, 2)
                plt.imshow(mask_to_show.squeeze(), cmap='gray')
                plt.title(f'Ground Truth Mask (Severity: {current_severity})')
                plt.axis('off')

                # Predicted binarized mask
                plt.subplot(1, 3, 3)
                plt.imshow(pred_mask_to_show.squeeze().numpy(), cmap='gray')
                plt.title('Predicted Mask (Binarized)')
                plt.axis('off')


                plt.tight_layout()
                plt.show()

                # Print the ground truth and predicted values
                print(f"Ground Truth Target: {target_label} (Score: {current_target}) | Ground Truth Severity: {current_severity}")
                print(f"Predicted Target: {predicted_target_label} (Score: {int(predicted_target_class)}) | Predicted Severity: {predicted_severity_score}")


                found_severities[current_severity] = True

        # Stop if we have found at least one example for each severity level
        if all(found_severities.values()):
            break

# Check if examples for all severity levels were found
if not all(found_severities.values()):
    missing_severities = [level for level, found in found_severities.items() if not found]
    print(f"Could not find examples for severity levels: {missing_severities} in the test set.")
```
